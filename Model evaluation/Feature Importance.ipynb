# Feature Importance

Feature importance helps identify which input variables (features) contribute the most to a model’s predictions.

It shows how much each feature influences the final output of the model.

## Why Feature Importance is Useful

- Helps understand the model better (interpretability)
- Identifies most impactful features
- Helps in feature selection (remove less important features)
- Improves model efficiency

## Key Idea

Not all features contribute equally.  
Some features have a strong impact on predictions, while others may have little or no effect.

## Example Question

Which features drive customer churn?

- High importance features (like contract type, monthly charges) influence churn more  
- Low importance features may have little or no impact  

## Conclusion

Feature importance helps us understand:

- What the model is learning  
- Which features matter most in predictions  

### Import libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Dataset
***Here, the breast cancer dataset from scikit-learn is used***

```python
data = load_breast_cancer()

X = data.data
y = data.target

feature_names = data.feature_names
```

### Train Model

```python
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
```

### Get Feature Importance

```python
importances = model.feature_importances_

feature_importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
})

feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

feature_importance_df.head(10)
```

### Visualize Feature Importance 

```python
plt.figure(figsize=(10,6))

plt.barh(feature_importance_df["Feature"][:10], feature_importance_df["Importance"][:10])
plt.gca().invert_yaxis()

plt.xlabel("Importance Score")
plt.title("Top 10 Important Features")
plt.show()
```

### What this shows
- Higher value → feature is more important
- Lower value → feature has less influence
- Helps understand what drives predictions
