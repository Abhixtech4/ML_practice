# A/B Testing

A/B Testing is used in production to compare two models and decide which one performs better.

- Model A → Current model  
- Model B → New model  

### Traffic Split
Users are randomly divided between both models:
- 50% → Model A  
- 50% → Model B  

This ensures a fair comparison.

### What to Measure
- **Conversion Rate** → How many users take desired action  
- **Accuracy** → Model prediction performance  
- **Revenue Impact** → Business outcome  

> Select the model that performs better based on real-world results.

### Import libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### Load dataset
***Here, the breast cancer dataset from scikit-learn is used***

```python
data = load_breast_cancer()
X, y = data.data, data.target
```

### Split data (simulate traffic split)

```python
X_A, X_B, y_A, y_B = train_test_split(
    X, y, test_size=0.5, random_state=42
)
```

### Model A (current)

```python
model_A = LogisticRegression(max_iter=5000)
model_A.fit(X_A, y_A)
pred_A = model_A.predict(X_B)
```

### Model B (new)

```python
model_B = RandomForestClassifier(random_state=42)
model_B.fit(X_B, y_B)
pred_B = model_B.predict(X_A)
```

### Evaluate

```python
acc_A = accuracy_score(y_B, pred_A)
acc_B = accuracy_score(y_A, pred_B)
print("Model A Accuracy:", acc_A)
print("Model B Accuracy:", acc_B)
```

### Comparison

```python
if acc_B > acc_A:
    print("\nModel B performs better → Replace Model A")
else:
    print("\nModel A performs better → Keep current model")
```
