# Handling Imbalanced Data

When one class has significantly more samples than another, models may become biased toward the majority class.

### Oversampling
Increase the number of samples in the minority class.

#### Why use it?
- Helps the model learn minority class patterns  
- Reduces bias toward majority class  

#### How it works:
- Duplicate minority class samples, or  
- Create synthetic samples (e.g., SMOTE)  

#### Example:
Fraud detection:
- 99% → Not Fraud  
- 1% → Fraud  
Oversampling increases fraud cases for better learning.

### Undersampling
Reduce the number of samples in the majority class.

#### Why use it?
- Useful when the dataset is very large  
- Reduces training time  

#### How it works:
- Randomly remove samples from the majority class  

#### Note:
- May lead to loss of important information if too much data is removed.

## 1. Oversampling

### Install imbalanced-learn if not already installed
```python
!pip install imbalanced-learn
```

### Import libraries

````python
from sklearn.datasets import fetch_openml
from collections import Counter
from imblearn.over_sampling import SMOTE
````

### Load dataset
***Here, the breast cancer dataset from scikit-learn is used***

````python
data = fetch_openml("creditcard", version=1, as_frame=True)
X = data.data
y = data.target.astype(int)
````

### Show original class distribution

````python
print("Original class distribution:")
print(Counter(y))
````

### Apply SMOTE

````python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
````

### Show new class distribution

````python
print("\nAfter applying SMOTE:")
print(Counter(y_resampled))
````

## 2. Undersampling

### Import libraries

````python
from sklearn.datasets import make_classification
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler
````

### Create an imbalanced dataset 

````python
X, y = make_classification(
    n_samples=20000,
    n_features=20,
    n_classes=2,
    weights=[0.95, 0.05],  
    random_state=42
)
````

### Show original class distribution

````python
print("Original class distribution:")
print(Counter(y))
````

### Apply undersampling

````python
undersample = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = undersample.fit_resample(X, y)
````

### Show new class distribution

````python
print("\nAfter applying Undersampling:")
print(Counter(y_resampled))
````
