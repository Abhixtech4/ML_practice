### Cross Validation

Cross Validation is a technique used to evaluate a machine learning model by training and testing it on multiple different splits of the dataset instead of a single train-test split.

### Why It Is Used
- Provides more reliable performance evaluation  
- Reduces dependency on one random split  
- Helps detect overfitting and underfitting  

### 1. K-Fold Cross Validation

The dataset is divided into **K equal parts (folds)**.

#### Process:
- Train on (K-1) folds  
- Test on the remaining 1 fold  
- Repeat K times with different test folds  
- Take the average performance score  

#### Example:
If K = 5:
- Model is trained and tested 5 times  
- Final result = average of all 5 scores  

### 2. Stratified K-Fold

Used when the dataset is imbalanced.

- Ensures each fold has the same class distribution  
- Prevents biased evaluation  

> Cross Validation gives a more stable and trustworthy estimate of model performance compared to a single train-test split.

---

## 1. K-Fold Cross Validation

### Import libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np
```

### Load dataset
***Here, the breast cancer dataset from scikit-learn is used***

```python
data = load_breast_cancer()
X, y = data.data, data.target
```

### Model

```python
model = LogisticRegression(max_iter=5000)
```

### K-Fold setup

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

### Cross validation scores

```python
scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
```

### Results

```python
print("Accuracy scores for each fold:", scores)
print("Mean Accuracy:", np.mean(scores))
print("Standard Deviation:", np.std(scores))
```

### 2. Stratified K-Fold

### Import libraries

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from collections import Counter
import numpy as np
```

### Load dataset
***Here, the credit card dataset from scikit-learn is used***

```python
data = fetch_openml("creditcard", version=1, as_frame=True)
X = data.data
y = data.target.astype(int)
```

### Show class imbalance

```python
print("Class distribution (Imbalanced dataset):")
print(Counter(y))
```

### Model

```python
model = LogisticRegression(max_iter=1000)
```

### Stratified K-Fold

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

### Cross-validation

```python
scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
```

### Results

```python
print("\nAccuracy scores for each fold:", scores)
print("Mean Accuracy:", np.mean(scores))
print("Standard Deviation:", np.std(scores))
```
