# Dimensionality Reduction

Dimensionality reduction is a technique used in machine learning to reduce the number of input features in a dataset while preserving the important information.

It is useful when a dataset has a large number of features, which can make models slow, complex, and harder to interpret.

## PCA (Principal Component Analysis)

PCA is one of the most commonly used dimensionality reduction techniques.  
It transforms the original features into a smaller set of new features called **principal components**, which still capture most of the important information.

## Purpose of PCA

- Reduce the number of features  
- Remove multicollinearity (high correlation between features)  
- Speed up model training  
- Improve data visualization in 2D or 3D  

## When to Use PCA

- When the dataset has many correlated features  
- For image data (high number of pixels/features)  
- For high-dimensional datasets where computation is expensive  

> Instead of working with all original features, PCA creates fewer new features that summarize the original data as efficiently as possible.

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Load dataset
***Here, the breast cancer dataset from scikit-learn is used***

```python
data = load_breast_cancer()
X = data.data
y = data.target
print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

### Standardize the data (important for PCA)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Apply PCA (reduce to 2 dimensions for visualization)

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print("Original shape:", X_scaled.shape)
print("Reduced shape:", X_pca.shape)
```

### Visualize the 2D projection

```python
plt.figure(figsize=(8,6))

plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1],
            label="Malignant", alpha=0.7)

plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1],
            label="Benign", alpha=0.7)

plt.title("PCA on Breast Cancer Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid()
plt.show()
```
