# K-Means Clustering

K-Means is an **unsupervised learning algorithm** used to group similar data points into clusters.

## Key Idea
The algorithm divides data into **K groups (clusters)** based on similarity, where each group has a center called a **centroid**.

## How It Works
1. Choose number of clusters (K)
2. Randomly initialize centroids
3. Assign each data point to the nearest centroid
4. Recalculate centroids based on assigned points
5. Repeat until centroids stop changing

## Important Features
- No labels are used (unsupervised learning)
- Groups data based on distance
- Sensitive to initial centroids and scaling

## When to Use
- Customer segmentation
- Pattern discovery
- Image compression
- Grouping similar data points

## Advantages
- Simple and fast
- Works well for well-separated clusters
- Easy to interpret

## Limitations
- Must choose K manually
- Sensitive to outliers
- Works poorly with non-spherical clusters

### Create dataset
***Here, the make_blobs dataset from scikit-learn is used***

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_samples=300,
    centers=4,
    n_features=2,
    cluster_std=1.0,
    random_state=42
)
```

### Apply K-Means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)
```

### Get cluster labels and centroids

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

### Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    c='red',
    s=200,
    marker='X',
    label='Centroids'
)
plt.title("K-Means Clustering on make_blobs Dataset")
plt.legend()
plt.show()
```
