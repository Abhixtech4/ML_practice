# Transformers / Deep Learning (NLP)

Transformers are modern deep learning models used in Natural Language Processing (NLP) to understand and generate human language.

They are more powerful than traditional machine learning models because they can capture context and meaning of words in a sentence.

### Key Idea
Transformers use a mechanism called "attention" to focus on important words in a sentence and understand relationships between them.

### Popular Transformer Models
- BERT → Used for understanding text (e.g., classification, question answering)
- GPT → Used for generating text (e.g., chatbots, content generation)
- RoBERTa → Improved version of BERT with better performance

### Why use Transformers?
- Understand context better than older models
- Handle large and complex text data
- State-of-the-art performance in many NLP tasks

### Applications
- Chatbots
- Language translation
- Question answering systems

## Transformers / Deep Learning Pipeline (NLP)

Transformers are deep learning models used to process and understand sequential data like text (and sometimes audio).

They are the foundation of modern NLP models like BERT and GPT.

### 1. Input Data (Text)
Raw text is given as input.

Example:
- Sentences
- Paragraphs
- Questions

### 2. Preprocessing (Tokenization)
Text is converted into smaller units called tokens.

Steps:
- Split text into tokens (words/subwords)
- Convert tokens into numerical IDs
- Add special tokens (CLS, SEP, etc.)

### 3. Embedding Layer
Tokens are converted into dense vectors.

These embeddings represent:
- Meaning of words
- Context in sentence

### 4. Transformer Architecture

Key component: **Attention Mechanism**

- Self-Attention → understands relationship between words
- Multi-head attention → captures different meanings in parallel

Popular models:
- BERT → understanding-based model
- GPT → generation-based model
- RoBERTa → improved BERT version

### 5. Model Processing

Transformer layers learn:
- Context of words
- Sentence relationships
- Long-range dependencies

### 6. Output Layer (Applications)

- Text classification
- Question answering
- Text generation
- Translation
- Chatbots

> Text → Tokenization → Embeddings → Attention → Deep Learning Layers → Output

### Load Dataset
***Here, the imdb dataset from datasets is used***

```python
import warnings
from datasets import load_dataset
dataset = load_dataset("imdb")
train_data = dataset["train"]
test_data = dataset["test"]
warnings.filterwarnings('ignore')
```

### Load Tokenizer (BERT)

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

### Tokenize Text

```python
def tokenize(example):
    return tokenizer(example["text"], truncation=True, padding="max_length")

train_data = train_data.map(tokenize, batched=True)
test_data = test_data.map(tokenize, batched=True)
```

### Load Model

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=2
)
```

### Training Setup

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=1
)
```

### Trainer

```python
from transformers import Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=test_data
)
trainer.train()
```

### Evaluation

```python
trainer.evaluate()
```

### Test with Custom Input

```python
inputs = tokenizer("This movie was amazing!", return_tensors="pt", truncation=True)
outputs = model(**inputs)
pred = outputs.logits.argmax().item()
print("Prediction:", "Positive" if pred == 1 else "Negative")
```
