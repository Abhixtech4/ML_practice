# Hierarchical Clustering

Hierarchical Clustering is an **unsupervised learning algorithm** used to group similar data points into clusters by building a hierarchy of clusters.

## Key Idea
Instead of choosing K in advance, it builds a **tree-like structure (dendrogram)** of clusters and then cuts it at a desired level.

## Types of Hierarchical Clustering
### 1. Agglomerative (Bottom-Up) 
- Start with each point as its own cluster
- Merge closest clusters step by step

### 2. Divisive (Top-Down)
- Start with one big cluster
- Split it into smaller clusters

## How It Works (Agglomerative)
1. Treat each data point as a single cluster  
2. Find closest clusters  
3. Merge them  
4. Repeat until one cluster remains  
5. Cut dendrogram to get final clusters  

## Distance Methods
- Single linkage (closest points)
- Complete linkage (farthest points)
- Average linkage
- Ward’s method (minimizes variance) 

## Important Features
- No need to predefine number of clusters (K)
- Produces dendrogram (tree structure)
- Works well for small datasets

## Advantages
- Easy to understand visually  
- No need to choose K initially  
- Gives hierarchical structure  

## Limitations
- Slow for large datasets  
- Sensitive to noise and outliers  
- Cannot undo merges once done  

## When to Use
- Small to medium datasets  
- When you want hierarchy of clusters  
- When K is unknown initially  

### Create dataset
***Here, the make_blobs dataset from scikit-learn is used***

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_samples=150,
    centers=3,
    n_features=2,
    cluster_std=0.8,
    random_state=42
)
```

### Plot dendrogram

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(8, 5))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrogram (Hierarchical Clustering)")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

### Apply Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
labels = model.fit_predict(X)
```

### Visualization of clusters

```python
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title("Hierarchical Clustering on make_blobs")
plt.show()
```
