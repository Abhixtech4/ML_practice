# Text Data

Natural Language Processing (NLP) is a field of machine learning that focuses on analyzing and understanding text data.

It helps computers read, interpret, and make decisions based on human language.

### Common Tasks in NLP
- Text classification
- Sentiment analysis
- Spam detection
- Language translation

### How it works
Text data is first converted into numerical form (like word counts or vectors), so machine learning models can process it.

### Common Models Used
- Naive Bayes → Simple and effective for text classification
- Logistic Regression
- Support Vector Machines (SVM)

### Example
Spam detection and sentiment analysis are common applications where NLP is used to classify text into categories.

---

## Text Data Pipeline

Natural Language Processing (NLP) is the process of enabling machines to understand, analyze, and work with human language (text data).

### 1. Input Data (Text)
Raw text data is collected from different sources.

Example:
- Emails
- Reviews
- Tweets
- Articles

### 2. Text Preprocessing
Text is cleaned and prepared for analysis:

- Convert to lowercase  
- Remove punctuation and special characters  
- Remove stopwords (e.g., "is", "the", "and")  
- Tokenization (split text into words)  
- Stemming/Lemmatization (reduce words to root form)

### 3. Feature Extraction
Text is converted into numerical form so machines can understand it:

Common methods:
- Bag of Words (Count-based representation)  
- TF-IDF (important word weighting)  
- Word embeddings (Word2Vec, GloVe)

### 4. Model Building
Machine learning or deep learning models are applied:

- Naive Bayes → fast text classification  
- Logistic Regression / SVM → strong traditional ML models  
- Deep Learning → RNN, LSTM, Transformers  

### 5. Training Process
Model learns patterns between:
- Input text features (X)
- Output labels (y)

### 6. Prediction
New text is:
- Preprocessed  
- Converted into features  
- Passed to trained model  
- Classified into a category  

### 7. Applications
- Spam detection  
- Sentiment analysis  
- Chatbots  
- Language translation  
- Topic classification  

> Text → Cleaning → Feature Extraction → Model → Prediction

---

### Load Dataset
***Here, the 20 new groups dataset from scikit-learn is used***

```python
from sklearn.datasets import fetch_20newsgroups
data = fetch_20newsgroups(subset='train')
X = data.data   # text data
y = data.target # labels
print("Sample text:\n", X[0])
```

### Text Preprocessing (Convert text → numbers)

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
X_vectorized = vectorizer.fit_transform(X)
```

### Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42
)
```

### Train Model (Naive Bayes)

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train, y_train)
```

### Prediction

```python
y_pred = model.predict(X_test)
```

### Evaluation

```python
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```

### Test with Custom Input

```python
sample = ["This computer has great graphics and fast processor"]
sample_vec = vectorizer.transform(sample)
prediction = model.predict(sample_vec)
print("Predicted Category:", data.target_names[prediction[0]])
```
