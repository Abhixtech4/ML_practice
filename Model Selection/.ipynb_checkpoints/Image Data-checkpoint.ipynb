# Image Data (Computer Vision)

Image data refers to visual data such as photos or frames that a machine learning model can analyze. Unlike text or numerical data, images are made of pixels.

To understand images, machine learning uses a branch called **Computer Vision**.

### Convolutional Neural Networks (CNN)

CNNs are deep learning models designed specifically for image processing. They automatically learn patterns like edges, shapes, and objects from images.

### Popular CNN Architectures
- **ResNet** → Uses skip connections to train very deep networks efficiently  
- **VGG** → Simple and deep architecture using small 3x3 filters  
- **EfficientNet** → Optimized model that balances accuracy and performance

### Common Applications
- Image classification → Identifying what is in an image  
- Object detection → Locating objects inside images  
- Face recognition → Identifying or verifying a person  

### Why CNNs are used?
- They work directly on pixel data  
- They automatically extract important features  
- They perform better than traditional ML for image tasks  

---

## Image Data Processing Pipeline (ML)

Image data (photos, frames, etc.) cannot be directly used by machine learning models. It must be processed into structured numerical form.

### 1. Raw Image Input
Images are made of pixels (height × width × channels).

Example:
- RGB image → 3 channels (Red, Green, Blue)
- Grayscale image → 1 channel

### 2. Preprocessing
Before feeding into a model:
- Resize image (e.g., 224×224, 32×32)
- Normalize pixel values (0–255 → 0–1)
- Optional: data augmentation (flip, rotate, zoom)

### 3. Feature Extraction

Unlike traditional ML, deep learning extracts features automatically.

Common representations:
- Edges
- Shapes
- Textures
- Object parts

### 4. Model Input

Processed images are passed to deep learning models:

- **CNN (Convolutional Neural Networks)** → main model for images  
- Advanced architectures:
  - ResNet → deep networks with skip connections  
  - VGG → simple stacked convolution layers  
  - EfficientNet → optimized performance vs accuracy  

### 5. Model Output (Applications)

- Image classification → identify object in image  
- Object detection → locate objects in image  
- Face recognition → identify individuals  
- Medical imaging → detect diseases  

> Image → Pixel processing → CNN → Feature learning → Prediction

---

### Load Dataset
***Here, the diabetes dataset from scikit-learn is used***

```python
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
print(X_train.shape, y_train.shape)
```

### Normalize Pixel Values (0 to 1)

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

### Define Class Names

```python
class_names = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
]
```

### Build CNN Model

```python
from tensorflow.keras import layers, models

model = models.Sequential()

# First Convolution Layer
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
model.add(layers.MaxPooling2D((2,2)))

# Second Convolution Layer
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))

# Third Convolution Layer
model.add(layers.Conv2D(64, (3,3), activation='relu'))

# Flatten + Dense Layers
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))

# Output Layer
model.add(layers.Dense(10, activation='softmax'))
```

### Compile Model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

### Train Model

```python
history = model.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test, y_test)
)
```

### Evaluate Model

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)
```

### Make Prediction

```python
import numpy as np

sample = X_test[0].reshape(1, 32, 32, 3)
prediction = model.predict(sample)

pred_class = np.argmax(prediction)

print("Predicted:", class_names[pred_class])
print("Actual:", class_names[y_test[0][0]])
```
