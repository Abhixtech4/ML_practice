# Gaussian Mixture Model (GMM)

Gaussian Mixture Model (GMM) is an **unsupervised learning algorithm** used for clustering. It assumes that data is made up of a mixture of several Gaussian (normal) distributions.

## Key Idea
Instead of assigning a data point to only one cluster (like K-Means), GMM gives a **probability of belonging to each cluster**.

## How It Works
1. Assume data is generated from multiple Gaussian distributions  
2. Each cluster has its own mean and variance  
3. Use **Expectation-Maximization (EM)** algorithm:
   - **E-step**: Estimate probabilities of cluster membership  
   - **M-step**: Update parameters (mean, variance)  
4. Repeat until convergence  

## Difference from K-Means
- K-Means → hard clustering (one cluster per point)  
- GMM → soft clustering (probabilities for each cluster)  

## Important Features
- Each cluster is modeled as a Gaussian distribution  
- Provides probabilistic assignments  
- Can handle overlapping clusters  

## Advantages
- More flexible than K-Means  
- Works well with overlapping data  
- Captures complex cluster shapes  

## Limitations
- More computationally expensive  
- Sensitive to initialization  
- Assumes data follows Gaussian distribution  

## When to Use
- When clusters overlap  
- When probabilistic clustering is needed  
- When data is not strictly spherical  

### Create dataset (with overlap for GMM)
***Here, the make_blobs dataset from scikit-learn is used***

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_samples=300,
    centers=3,
    n_features=2,
    cluster_std=2.5,  # overlap between clusters
    random_state=42
)
```

### Train Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)
```

### Predict cluster labels

```python
labels = gmm.predict(X)
```

### Get cluster probabilities (soft clustering)

```python
probs = gmm.predict_proba(X)
```

### Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(
    gmm.means_[:, 0],
    gmm.means_[:, 1],
    c='red',
    s=200,
    marker='X',
    label='Cluster Centers'
)

plt.title("Gaussian Mixture Model (GMM) on make_blobs Dataset")
plt.legend()
plt.show()
```
