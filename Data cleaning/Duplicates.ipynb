# Duplicates in Dataset

## What are Duplicates?
Duplicates are rows in a dataset that have exactly the same values across all columns (or sometimes specific columns). These repeated records do not add new information.

## Why Duplicates are a Problem?
- They can **bias the model** by over-representing certain data.
- They may lead to **incorrect insights** during analysis.
- They can **increase dataset size unnecessarily**, slowing down processing.

## When Do Duplicates Occur?
- Data entry mistakes
- Merging multiple datasets
- System or logging errors
- Web scraping issues

## How to Handle Duplicates?
1. **Detect duplicates** → Identify repeated rows  
2. **Analyze them** → Decide if they are truly unwanted  
3. **Remove or keep** → Based on context and need  

## When Should You Remove Duplicates?
- When duplicates are **unintentional errors**
- When they **distort analysis or model training**

## When Should You Keep Duplicates?
- When duplicates represent **real repeated events** (e.g., transactions, clicks)
- When frequency of occurrence is **important information**

> Not all duplicates are bad — always understand your data before removing them.

## Dataset Used

We are using the Titanic dataset available in seaborn

This dataset contains information about passengers such as age, gender, class, and whether they survived or not.

```python
import seaborn as sns
df = sns.load_dataset("titanic")
```

---

## Checking Number of Duplicate Rows

```python
df.duplicated().sum()
```

### Explanation
- `df.duplicated()` → Checks each row and returns **True** if the row is a duplicate, otherwise **False**
- `.sum()` → Counts all **True values** (since True = 1, False = 0)
- The dataset contains **107 duplicate rows**
- `np.int64` means the result is an integer type from NumPy

### When to Use This?
- At the **initial stage of data cleaning**
- To quickly check if your dataset has duplicate entries

---

## Removing Duplicate Rows

```python
df.drop_duplicates()
```

### Explanation:
- Removes all **duplicate rows** from the dataset  
- Keeps only the **first occurrence** of each duplicate row  
- Returns a **new DataFrame** (original data remains unchanged)  
- The dataset now contains **only unique rows**  
- Duplicate rows have been removed  
- Final shape: **784 rows × 15 columns**  
- This means **107 duplicate rows** were removed from the original dataset  

### When to Use This?
- After detecting duplicates in your dataset  
- When duplicate rows are **errors or unnecessary**  
- Before training ML models to avoid bias
