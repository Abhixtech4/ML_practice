# Data Type Correction 

Data type correction means making sure each column in your dataset has the correct data type (such as integer, float, string, or date). Sometimes, data is stored in the wrong format—for example, numbers saved as text or dates stored as strings. This can lead to errors or incorrect results during analysis and machine learning. Therefore, we convert each column to the appropriate data type to ensure the data is accurate, consistent, and ready for further processing.

---

### Loading Dataset

```python
import pandas as pd
df = pd.read_csv("my_netflix_dataset.csv")
```

In this step, we import the pandas library and load the dataset using `pd.read_csv()`. This function reads the CSV file and converts it into a DataFrame, which is easier to work with for analysis and data cleaning.

The dataset used here is taken from Kaggle:  
https://www.kaggle.com/datasets/pulkitchhabra12/netflix-titles-csv

---

### Understanding Data Types

```python
df.info()
```

We use `df.info()` to get a summary of the dataset. It shows the number of rows and columns, column names, non-null values, and data types.

From the output, most columns have the `object` data type (text data), while `release_year` is in `int64` (numeric format).  

This step helps us identify which columns may need data type correction, such as converting dates or numerical values stored as text into proper formats.

---

### Converting Date Column

```python
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"])
```

In this step, we clean and convert the `date_added` column to the correct data type.

First, `.str.strip()` is used to remove any extra spaces from the text.  
Then, `pd.to_datetime()` converts the column from string (object) to datetime format.

This ensures the column is properly formatted for date-based analysis.

---

### Converting Numeric Column

```python
df["release_year"] = df["release_year"].astype(int)
```

In this step, we convert the `release_year` column to integer type using `.astype(int)`.

This ensures the column is in the correct numeric format, making it suitable for calculations and analysis.

---

### Extracting Numeric Values from Text

```python
df["duration_num"] = df["duration"].str.extract(r'(\d+)').astype(float)
```

In this step, we create a new column `duration_num` by extracting numeric values from the `duration` column.

The `.str.extract(r'(\d+)')` function extracts only the numbers from the text (e.g., "90 min" → 90).  
Then, `.astype(float)` converts these values into a numeric (float) format.

This makes the duration easier to use for analysis and calculations.

---

### Updated Data Types After Correction

```python
df.info()
```

After performing data type corrections, we use `df.info()` again to verify the changes.

Now, `date_added` is converted to `datetime64[ns]`, and the new column `duration_num` is in `float64` format. The `release_year` remains in `int64`.

This confirms that the dataset now has more appropriate data types, making it cleaner and better for analysis and machine learning.
