# Handling Inconsistent Values

Inconsistent values occur when the same category is written in different formats, such as "Male", "male", "M", or "m". This can create problems during analysis because they are treated as different values.

To fix this, we standardize the data:
- Convert all text to lowercase using `.str.lower()`
- Replace inconsistent labels with a common format using `.replace()`

This ensures uniformity in the dataset and improves the quality of analysis.

---

### Loading Dataset

```python
import pandas as pd
df = pd.read_csv("adult.csv")
```

In this step, we import the pandas library and load the dataset using `pd.read_csv()`. This function reads the CSV file and converts it into a DataFrame for easy analysis.

The dataset used here is the Adult Census Income dataset, taken from Kaggle:  
https://www.kaggle.com/datasets/lovishbansal123/adult-census-income

This dataset contains demographic and employment-related information, and it is commonly used to predict whether a person's income is greater than 50K per year.

---

### Previewing the Dataset

```python
df.head()
```

We use `df.head()` to display the first few rows of the dataset. This helps us quickly inspect the data, understand its structure, and identify potential issues like missing or inconsistent values.

From the output, we can see that some columns (like `workclass` and `occupation`) contain `"?"`, which may indicate missing or inconsistent data. Columns like `sex` have values such as `"Female"` that might need standardization.

---

### Standardizing Gender Values

```python
df["sex"] = df["sex"].str.lower().str.strip()
df["sex"] = df["sex"].replace({"m": "male", "f": "female"})
```

In this step, we clean the `sex` column to fix inconsistent values:

1. `.str.lower().str.strip()` converts all text to lowercase and removes extra spaces.  
2. `.replace({"m": "male", "f": "female"})` replaces shorthand labels with the standard format.

This ensures that all gender values are consistent, making analysis and modeling more accurate.
