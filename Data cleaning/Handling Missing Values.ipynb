# Handling Missing Values

Missing values are common in real-world datasets. Handling them properly is important because most machine learning algorithms cannot handle `NaN` or empty values directly.

## Why Handle Missing Values?
- Prevent errors during model training.
- Avoid biased or inaccurate results.
- Improve model performance and reliability.

## Common Strategies
1. **Removing missing values**  
   - Drop rows or columns containing missing data.  
   - Useful when missing data is small or irrelevant.

2. **Imputation**  
   - Fill missing values using:  
     - **Mean / Median / Mode** for numerical data.  
     - **Most frequent / constant** for categorical data.  
   - Preserves dataset size.

3. **Prediction-based methods**  
   - Predict missing values using other features.  
   - Advanced but more accurate.

4. **Flagging missing values**  
   - Create a new column to indicate missing values.  
   - Helps model learn patterns related to missing data.

**Use Case:** This is used in **data cleaning, preprocessing, and preparation** before feeding data into any ML model.

# 1. Checking for missing value

## inspect missing data

````python
df.isnull().sum()
````

### Explanation
- `df.isnull().sum()` counts the missing values (`NaN`) in each column.  
- Columns like `age`, `deck`, and `embarked` have missing values.  
- Columns with `0` missing values are complete and do not need handling.  

### Use for
- Use this method to **quickly inspect missing data** in any dataset before deciding on imputation or removal strategies.

---

## checking missing values as percentage

````python
df.isnull().mean() * 100
````

### Explanation
- `df.isnull().mean() * 100` calculates the **percentage of missing values** for each column.  
- `age` has ~20% missing values, `deck` has ~77%, while columns like `survived` have 0%.  
- High missing percentage (like `deck`) may need careful handling; sometimes dropping the column is better.  

### Use for
- Use this method to **prioritize columns for imputation or removal** based on how much data is missing.

# 2. Handling missing value

## Dropping Columns with High Missing Values

````python
df.drop(columns=["deck"])
````

Sometimes, columns with a very high percentage of missing data can be **dropped** instead of imputing them.

### Explanation
- We removed the `deck` column because it has **more than 77% missing values**, making imputation less practical.  
- **Caution:** `df.drop()` does **not permanently remove** the column.  
- To permanently drop the column, use:  
```python
df.drop(columns=["deck"], inplace=True)
```
### Use for
- Drop columns with very high missing values to simplify the dataset and avoid unnecessary imputation.

---

## Handling Missing Values by Imputation

````python
df["age"].fillna(df["age"].median())
````

For columns with missing values, we can **fill them using statistical measures** like mean, median, or mode.  

### Explanation
- `df["age"].fillna(df["age"].median())` replaces all missing values in the `age` column with the **median age**.  
- Median is used instead of mean when the data may have **outliers**.  
- **Caution:** This does **not modify the original dataframe** unless you use `inplace=True` or assign it back:  
```python
df["age"] = df["age"].fillna(df["age"].median())
```

### Use for
- Use imputation to handle missing numerical values without losing data by dropping rows.
- Median is especially useful when the column has outliers.

---

## Handling Missing Categorical Values by Mode Imputation

````python
df["embarked"].fillna(df["embarked"].mode()[0])
````

For categorical columns with missing values, we can **fill them using the most frequent value (mode)**.

### Explanation
- `df["embarked"].fillna(df["embarked"].mode()[0])` replaces all missing values in the `embarked` column with the **most frequent value**.  
- Mode imputation is suitable for **categorical columns**. General Use
Use mode imputation to handle missing categorical values without dropping rows.
Useful when the missing data is small and you want to preserve the dataset. 
- **Caution:** This does **not modify the original dataframe** unless you use `inplace=True` or assign it back:  
```python
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
```
### Use for
- Use mode imputation to handle missing categorical values without dropping rows.
- Useful when the missing data is small and you want to preserve the dataset.

---

## Handling Missing Values in Numerical Columns by Median Imputation

````python
df["fare"].fillna(df["fare"].median())
````

 numerical columns with missing values, we can **fill them using median** to avoid losing data.

### Explanation
- `df["fare"].fillna(df["fare"].median())` replaces all missing values in the `fare` column with the **median fare**.  
- Median is preferred when the column may have **outliers**.  
- **Caution:** This does **not modify the original dataframe** unless you use `inplace=True` or assign it back:  
```python
df["fare"] = df["fare"].fillna(df["fare"].median())
```
### Use for
- Use median imputation to handle missing numerical values without dropping rows.
- Helps maintain dataset size and reduces bias caused by outliers.

---

## Handling Missing Values Using Forward Fill (ffill)

````python
df.ffill()
````

Forward fill propagates the **last valid observation forward** to fill missing values.

### Explanation
- `df.ffill()` fills missing values by **carrying the previous row’s value forward**.  
- Useful when data is **ordered** (e.g., time series) and previous values are reasonable replacements.  
- **Caution:** This does **not modify the original dataframe** unless you use `inplace=True` or assign it back:  
```python
df = df.ffill()
```
### Use for
- Use forward fill to fill missing data sequentially, especially in time-dependent datasets.
- Helps preserve dataset size without dropping rows.

---

## Handling Missing Values Using Backward Fill (bfill)

````python
df.bfill()
````

Backward fill fills missing values by **propagating the next valid observation backward**.

### Explanation
- `df.bfill()` fills missing values by **using the next row’s value**.  
- Useful when data is **ordered** and the next observation is a reasonable replacement.  
- **Caution:** This does **not modify the original dataframe** unless you use `inplace=True` or assign it back:  
```python
df = df.bfill()
```
### Use for
- Use backward fill to fill missing data sequentially from future values, especially in time-dependent datasets.
- Helps maintain dataset size without dropping rows.

## Handling Missing Values Using Machine Learning Model

````python
from sklearn.ensemble import RandomForestRegressor

# Use only relevant features
age_df = df[["age", "pclass", "sibsp", "parch", "fare"]]

# Split into known and unknown
known_age = age_df[age_df["age"].notnull()]
unknown_age = age_df[age_df["age"].isnull()]

X_train = known_age.drop("age", axis=1)
y_train = known_age["age"]

X_test = unknown_age.drop("age", axis=1)

# Train model
model = RandomForestRegressor()
model.fit(X_train, y_train)

# Predict missing Age
predicted_ages = model.predict(X_test)

# Fill back
df.loc[df["age"].isnull(), "age"] = predicted_ages

df['age'].isnull().sum()
````

For numerical columns, missing values can be **predicted using a machine learning model** based on other related features.

### Explanation
- We use a `RandomForestRegressor` to predict missing `age` values using other features.  
- After prediction, all missing `age` values are filled (`df['age'].isnull().sum()` returns `0`).  
- Model-based imputation is more accurate than simple mean/median when other features correlate with the missing column.

### Use for
- Use machine learning-based imputation for **numerical columns with patterns related to other features**.  
- Ideal for datasets where missing data is significant but predictable using other columns.
