# Basic Overview of Machine Learning Lifecycle

Machine Learning (ML) is the process of building systems that can learn patterns from data and make decisions or predictions.

The ML lifecycle is a step-by-step process that helps in building a successful ML model. It includes:

* Understanding the problem
* Collecting and exploring data
* Cleaning and preparing data
* Building and training the model
* Evaluating performance
* Making predictions and improving the model

This notebook gives a simple overview of these steps using an example dataset (Titanic dataset).

## Statistical Summary of Dataset

```python
df.describe()
```

### Explanation

* `df.describe()` gives statistical summary of all **numerical columns**.
* It helps to understand data distribution, central tendency, and spread.


### Output Understanding (How to Read)

Each column (like `age`, `fare`, etc.) has these values:

#### 1. **count**

* Number of non-missing values
* Example:

  * `age = 714` → 714 values present, rest are missing


#### 2. **mean (Average)**

* Sum of all values ÷ total values
* Example:

  * `age mean = 29.69` → average age is around 29 years


#### 3. **std (Standard Deviation)**

* Shows how spread out the data is from the mean
* High value → data is more spread
* Low value → data is closer to mean


#### 4. **min (Minimum)**

* Smallest value in the column
* Example:

  * `age min = 0.42` → youngest passenger


#### 5. **25% (1st Quartile)**

* 25% of data is below this value
* Example:

  * `age = 20.12` → 25% passengers are younger than ~20


#### 6. **50% (Median)**

* Middle value of the data
* 50% values are below and 50% are above
* Example:

  * `age = 28` → half passengers are younger than 28

👉 This is also called **median** (very important, especially when data has outliers)


#### 7. **75% (3rd Quartile)**

* 75% of data is below this value
* Example:

  * `age = 38` → most passengers are younger than 38


#### 8. **max (Maximum)**

* Largest value in the column
* Example:

  * `fare max = 512.32` → highest ticket price


### Key Insights from This Output

* `age` has missing values (count < 891)
* `fare` has high variation (large std) → possible outliers
* `sibsp` and `parch` mostly have low values (many zeros)
* Median (50%) is often better than mean when data is skewed


### Simple Understanding

* **mean** → average
* **50% (median)** → middle value
* **std** → spread of data
* **min/max** → range
* **25% & 75%** → distribution of data

## Check Missing Values

```python
df.isnull().sum() 
```

### Explanation

* `df.isnull()` checks for missing values (NaN) in the dataset.
* `.sum()` adds them column-wise.
* So, this code shows total missing values in each column.

### Output Understanding

* Each value represents number of missing values in that column.

#### Important Observations:

* `age = 177`
  → 177 values are missing in age column

* `deck = 688`
  → Large number of missing values (most data is missing)

* `embarked = 2` and `embark_town = 2`
  → Very few missing values

* Other columns = 0
  → No missing values

### Simple Understanding

* **0** → No missing data
* **High number** → Many missing values (needs attention)


### Why This is Important

* Missing data can affect model performance
* We may need to:

  * Fill missing values (imputation)
  * Drop columns (like `deck` if too many missing)
