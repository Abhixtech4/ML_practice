## Feature Understanding

Feature understanding is the first step in data analysis.

It helps you explore each feature (column) to understand:

- What type of data it contains (categorical or numerical)
- How values are distributed
- How features relate to each other and the target variable

This step is important for making decisions about preprocessing, feature selection, and modeling.

---

### Dataset Used

For this analysis, we use the **tips dataset** available from the Seaborn library.

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

---

```python
## Understanding data distribution
```

```python
df["day"].value_counts()
```

**Explanation:**  
Returns the frequency of each unique value in the selected column.

- **Index** → unique values from the feature  
- **Values** → count (frequency) of each unique value  
- Sorted in descending order by default  

**Use for:**  
Used to understand how data is distributed across categories.

---

## Identify possible value

```python
df["sex"].unique()
```

**Explanation:**  
Returns all unique values present in the selected column.

- Displays distinct values in the feature  
- May also show category information if the column is categorical  

**Use for:**  
Used to quickly identify possible values in a feature.

---

## Measure cardinality

```python
df["smoker"].nunique()
```

**Explanation:**  
Returns the number of unique values in the selected column.
 
- A single integer value  
- Represents the count of distinct values  

**Use for:**  
Used to measure cardinality of a feature.

---

## Identify relationships between numerical features

```python
df.corr(numeric_only=True)
```

**Explanation:**  
Computes pairwise correlation between numerical features in the dataset.

- Returns a correlation matrix  
- Values range between -1 and 1  
- 1 → strong positive correlation  
- -1 → strong negative correlation  
- 0 → no correlation  
- Diagonal values are always 1 (self-correlation)  

**Use for:**  
Used to identify relationships between numerical features and detect multicollinearity.

---

## Most related feature to target

```python
df.corr(numeric_only=True)["tip"]
```

**Explanation:**  
Shows correlation of all numerical features with a specific target feature (`tip` in this case).

- Index → numerical features  
- Values → correlation with the target feature  
- 1 → perfect positive correlation with itself  
- Positive/negative values indicate direction of relationship  

**Use for:**  
Used to identify which numerical features are most related to the target for feature selection.
