## Target Variable Analysis

Target variable analysis is one of the first steps in data understanding.  
It helps us understand the nature and distribution of the variable we want to predict.

- In **classification**, we check how classes are distributed.
- In **regression**, we study statistical properties like mean, spread, and skewness.

This step is important because:
- It helps detect class imbalance.
- It gives an idea about data distribution.
- It guides model selection and evaluation strategy.
---

### Dataset Used

For this analysis, we use the **tips dataset** available from the Seaborn library.

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

---

### Target Variable Count 

```python
df["tip"].value_counts()
```

**Explanation:**
- This shows how many times each tip value appears in the dataset.
- Tip = 2.00 appears 33 times (most common).
- Tip = 3.00 appears 23 times.
- Some tip values appear only once (rare cases).
- Total unique tip values = 123.

**This shows:**
- `value_counts()` counts frequency of each unique value in a column.
- Helps to understand:
  - Most frequent values
  - Rare values
  - Distribution of data
- Useful for checking imbalance and variability in target variable.

### Target Variable Distribution (Percentage)

```python
df["tip"].value_counts(normalize=True)
```

**Explanation :**
- This shows the proportion (percentage) of each tip value.
- Tip = 2.00 makes up about 13.5% of total data (most common).
- Tip = 3.00 makes up about 9.4%.
- Rare values have very small proportions (~0.4%).
- Total of all values = 1 (100%).

**This shows:**
- `value_counts(normalize=True)` gives relative frequency instead of count.
- Helps to understand:
  - Percentage distribution of values
  - Dominant vs rare values
- Useful for comparing importance of values and detecting imbalance.

## Statastical summary for target variable

```python
df["tip"].describe()
```

**Explanation:**
- Total values = 244.
- Average (mean) tip ≈ 3.00.
- Standard deviation ≈ 1.38 → moderate spread.
- Minimum tip = 1.00, Maximum tip = 10.00.
- 25% values ≤ 2.00 (lower quartile).
- 50% values ≤ 2.90 (median).
- 75% values ≤ 3.56 (upper quartile).

**This shows:**
- `describe()` gives statistical summary of numerical data.
- Includes:
  - Count, Mean, Standard Deviation
  - Min, Max
  - Quartiles (25%, 50%, 75%)
- Helps to understand:
  - Central tendency (mean, median)
  - Spread of data
  - Range and possible outliers

## checking skewness in target variable.

```python
df["tip"].skew()
```

**Explanation:**
- Skewness ≈ 1.47 (positive value).
- Data is right-skewed (long tail on the right side).
- Most tip values are smaller, with few large values.

**This shows:**
- Skewness measures asymmetry of data distribution.
- If:
  - Skew = 0 → symmetric data
  - Skew > 0 → right-skewed (more small values, few large)
  - Skew < 0 → left-skewed (more large values, few small)
- Helps to understand distribution shape and need for transformation.
