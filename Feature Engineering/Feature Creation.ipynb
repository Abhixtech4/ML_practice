# Feature Creation (Feature Engineering)

Feature creation is the process of generating new input variables (features) from existing data to improve the performance of a machine learning model.

Instead of using raw data directly, we transform or combine columns to create more meaningful information.

### Why Feature Creation is Important?
- Helps models understand patterns better
- Improves accuracy and performance
- Converts raw data into useful signals

> Feature creation helps turn raw data into meaningful features that make machine learning models more effective.

---

## 1. Extracting Date Features (Customer Transactions Dataset)

### Load datset
datset link: https://www.kaggle.com/datasets/fares279/customers-transactions?utm_source=chatgpt.com

```python
import pandas as pd
df = pd.read_csv("Customer_Transactions.csv")
```

### Reading first five row

```python
df.head()
```

###  creating Feature

```python
df["last_purchase_date"] = pd.to_datetime(df["last_purchase_date"])
df["Year"] = df["last_purchase_date"].dt.year
df["Month"] = df["last_purchase_date"].dt.month
df["Day"] = df["last_purchase_date"].dt.day
```

### again read dataset after creating feature

```python
df.head()
```

In this example, we use the **Customer Transactions Dataset** from Kaggle.

The dataset contains a column `last_purchase_date`, which we convert into a datetime format and extract useful components like year, month, and day.

This helps the model capture time-based patterns in customer behavior.

### Why this is useful?
- Identifies seasonal trends (month)
- Captures yearly patterns
- Helps analyze customer activity over time

## 2. Profit Calculation & Date Feature Extraction 

### Load datset
dataset link: https://www.kaggle.com/datasets/faresashraf1001/supermarket-sales

```python
import pandas as pd
df = pd.read_csv("SuperMarket Analysis.csv")
```

### Read first five row

```python
df.head()
```

### Creating profit feature in dataset

```python
df["Profit"] = df["Sales"] - df["cogs"]
```

### Creating year feature from the date
**Note:** in dataset date is not in datetime format so first convert it into datetime format and then creating feature.

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Year"] = df["Date"].dt.year
```

### Read dataset after creating feature

```python
df.head()
```

In this example, we use the **Supermarket Sales Dataset** from Kaggle.

We create a new feature called `Profit` using sales and cost data. We also convert the `Date` column into datetime format and extract the year.

### Why this is useful?
- Profit helps measure business performance
- Year helps identify trends over time
- Improves model understanding with meaningful features

## 3 Creating Age from Birth Date

### Load datset
dataset link: https://www.kaggle.com/datasets/mexwell/famous-birthdays

```python
import pandas as pf
df = pd.read_csv("birthdays.csv")
```

### Read first five row

```python
df.head()
```

### Creating age feature

**Note:**
- Firstly correct the format of birthdate to datetime
- Importing datetime from datetime modile which used to get today's date
- After that applying simple math, convert it into age

```python
from datetime import datetime
df["birthDate"] = pd.to_datetime(df["birthDate"], errors="coerce")
today = datetime.today()
df["Age"] = today.year - df["birthDate"].dt.year
```

### Read dataset after creating feature

```python
df.head()
```

In this example, we use the **Famous Birthdays Dataset** from Kaggle.

We convert the `birthDate` column into datetime format and create a new feature `Age` using the current year.

### Why this is useful?
- Converts date into a numerical feature (age)
- Makes data more useful for modeling
- Helps analyze age-related patterns
