# **Encoding Categorical Variables**

In machine learning, **Encoding** is the process of converting categorical data (text-based labels like "Red," "Green," or "Large") into **numerical values**.

#### **Why do we need it?**
Most machine learning algorithms are based on mathematical equations. Since you cannot perform math on a string of text, we must translate those labels into numbers so the model can process the features and find patterns.

#### **The Core Goal**
The objective is to represent the data numerically while preserving its meaning. We want to ensure the model understands:
* Whether the categories have a specific **rank/order** (Ordinal).
* Whether the categories are **independent** of each other (Nominal).

---

## 1. Label Encoding

Converts categorical values into numbers while preserving order.

**When to use:**
Use when data has a natural order (ordinal data).

**Dataset used:**
Students Performance in Exams Kaggle dataset
https://www.kaggle.com/code/pragya1110/student-performance-in-exam?select=StudentsPerformance.csv

**Column:**
parental level of education

**Why use here:**
This column has order (high school < ... < master’s), so label encoding keeps this ranking, helping the model understand progression.

**Example:**
high school → 0, …, master’s → 4


### Load Dataset

```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
```

### Read first five rows

```python
df.head()
```

### Find unique value in parental level of education

```python
df["parental level of education"].unique()
```

### Do label encoding bases on education level

```python
education_order = [
    "some high school",
    "high school",
    "some college",
    "associate's degree",
    "bachelor's degree",
    "master's degree"
]

mapping = {k: i for i, k in enumerate(education_order)}

df["parent_edu_encoded"] = df["parental level of education"].map(mapping)

print(df[["parental level of education", "parent_edu_encoded"]].head())
```

## 2. One-Hot Encoding

Used when there is no order between categories (nominal data).

**When to use:**
Use when categorical values are independent and have no ranking.

**Dataset used:**
Seaborn Titanic dataset

**Column:**
Embarked

**Why use here:**
Embarked has categories (C, Q, S) with no order or priority. So One-Hot Encoding is used to avoid false ranking that would happen in label encoding.

**Example:**
C → [1,0,0]

Q → [0,1,0]

S → [0,0,1]

### Load dataset

```python
import seaborn as sns
df = sns.load_dataset("titanic")
```

### Read first five rows

```python
df.head()
```

### Do One-Hot encoding

```python
df_encoded = pd.get_dummies(df, columns=["embarked"], drop_first=False)
print(df_encoded[["embarked_C", "embarked_Q", "embarked_S"]].head())
```

## 3. Target Encoding

Replace each category with the mean of the target variable for that category.

**When to use:**
Use when categorical variable has strong relationship with the target and high cardinality.

**Dataset used:**
Titanic dataset

**Column:**
Embarked

**Why use here:**
Embarked categories (C, Q, S) are replaced with the mean survival rate of each group. This helps the model capture the relationship between boarding location and survival.

**Example:**
C → mean survival of C passengers
Q → mean survival of Q passengers
S → mean survival of S passengers

### Load dataset

```python
import seaborn as sns
df = sns.load_dataset("titanic")
```

### Read first five rows

```python
df.head()
```

### Do target encoding

```python
target_mean = df.groupby("embarked")["survived"].mean()
df["embarked_target_encoded"] = df["embarked"].map(target_mean)
print(df[["embarked", "embarked_target_encoded"]].head())
```
