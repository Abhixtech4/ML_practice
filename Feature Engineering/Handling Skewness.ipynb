## Handling Skewness

### What is Skewness?
Skewness is the measure of asymmetry in data distribution.

- **Positive skew:** Tail on the right  
- **Negative skew:** Tail on the left  

### Why it matters
Skewed data can affect model performance and reduce accuracy, especially in models that assume normal distribution.

### Check skewness
```python
df['column'].skew()
```

- 0 → Positive skew
- < 0 → Negative skew
- ≈ 0 → Normal distribution

> Use transformations to reduce skewness and make data more suitable for machine learning models.

---

### Load dataset 
dataset used for meassuring skewness is taken from seaborn

````python
import seaborn as sns
df = sns.load_dataset("diamonds")
````

### Checking skewness on price column

````python
df['price'].skew()
````

---

## 1. Log Transformation for Handling Skewness

````python
import numpy as np
np.log1p(df['price'])
````

Log transformation is used to reduce **positive (right) skewness** in data. It compresses large values and makes the distribution more normal-like.

### When to use it
- Data is positively skewed  
- Data contains large outliers
  
`log1p()` is used instead of `log()` because it safely handles zero values.

### Making changes permanent
``` python
df['price'] = np.log1p(df['price'])
```
This updates the column in the dataset.

(Note: pandas does not support `inplace=True` for this operation directly.)

> Log transformation helps reduce skewness by compressing large values and improving data distribution for machine learning models.

---

## 2. Square Root Transformation for Handling Skewness

````python
np.sqrt(df['price'])
````

Square root transformation is used to reduce **moderate positive skewness** in data. It reduces the impact of large values while keeping the data easy to interpret.

### When to use it
- Data is moderately positively skewed  
- Data does not have extreme outliers  

### Make changes permanent
``` python
df['price'] = np.sqrt(df['price'])
```
This updates the dataset column directly.

> Square root transformation helps reduce skewness by compressing higher values and making the distribution more balanced.

---

## 3. Box-Cox Transformation for Handling Skewness

````python
from scipy import stats
stats.boxcox(df['price'])
````

Box-Cox transformation is used to reduce **positive skewness** and make data more normally distributed. It is more powerful than log or square root transformation.

### When to use it
- Data is positively skewed  
- Data has only **positive values** (no zero or negative values)  

### Make changes permanent
``` python
df['price'] = stats.boxcox(df['price'])
```
This updates the dataset column directly.

> Box-Cox transformation reduces skewness and improves normality, making data more suitable for machine learning models.

---

## 4. Yeo-Johnson Transformation for Handling Skewness

````python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
pt.fit_transform(df[['price']])
````

````python
Yeo-Johnson transformation is used to reduce skewness and make data more normally distributed. It works like Box-Cox but can handle **zero and negative values** as well.

### When to use it
- Data is skewed (positive or negative)
- Data contains zero or negative value

### Make changes permanent
``` 
df[['price']] = pt.fit_transform(df[['price']])

This updates the dataset column directly.

Summary

Yeo-Johnson transformation reduces skewness and is more flexible than Box-Cox because it works with zero and negative values.s

````
